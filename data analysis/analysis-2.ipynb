## Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
input_path = '../input/'

train = pd.read_csv(input_path + 'train.csv')
test = pd.read_csv(input_path + 'test.csv')
```

```python
train.head(3)
```

```python
# 성별 생존율
live = train[train['Survived']==1]['Sex'].value_counts()
dead = train[train['Survived']==0]['Sex'].value_counts()

df = pd.DataFrame([live, dead])
df.index = ['Live', 'Dead']
df.plot.bar(stacked=True, rot=45)
plt.show()
```

```python
# bar chart 함수
def bar_chart(feature):
    live = train[train['Survived']==1][feature].value_counts()
    dead = train[train['Survived']==0][feature].value_counts()
    df = pd.DataFrame([live, dead])
    df.index = ['Live', 'Dead']
    df.plot.bar(stacked=True, rot=45)
    plt.show()
```

```python
bar_chart('Pclass')
```

```python
bar_chart('SibSp')
```

```python
bar_chart('Parch')
```

```python
bar_chart('Embarked')
```
